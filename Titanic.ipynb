```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
sns.countplot(x="Survived", data = df, hue = "Survived", palette = "Set2", legend = False)
plt.title("Survival Count 0 = Not Survived, 1 = Survived")
plt.show()
```

```python
sns.countplot(x = "Sex", data = df, hue = "Sex", palette = "Set2", legend = False)
plt.title("Passenger Gender Distribution")
plt.show()
```

```python
sns.histplot(df["Age"].dropna(), bins = 30, kde = True, color = "skyblue")
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x = "Sex", data = df, hue = "Sex", palette = "Set1")
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df, palette="Set2")
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[["Survived", "Age", "Fare", "Pclass"]], hue="Survived", palette="husl")
plt.show()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.drop(columns=["Pclass"], inplace=True)

print("\nAfter cleaning missing values:\n", df.isnull().sum())
```
